```python
#!/usr/bin/env python3
# Name: Toby Grief
# Group Members: Nathan Perez, Theodore Shen, Bharath Karintha, Nate Cruz
'''
myBio
Print your:
    - name
    - username
    - student status
    - major
    - why you are taking the class and hopes
    - MolBio problem that interests you
    - prio computer science exp
Do so by defining a Person class
input: strings of arbitrary lengths that will define the new person object
output: greeting is printed on screen

'''
class Person:
    '''This code builds a new Person object with the assigned attributes when instantiated'''
    def __init__(self,name,username,major,studentType,why,MolBio,PriorCompExp):
        '''This saves all instances arguments'''
        self.myName = name
        self.myUsername = username
        self.myMajor = major
        self.myStudentType = studentType
        self.myWhy = why
        self.myMolBio = MolBio
        self.PriorCompExp = PriorCompExp

    def introduce (self):
        '''This prints the person's attributes'''
        print("Hi there, my name's {0}. \nMy username is {1}.".format(self.myName,self.myUsername))
        print('I am a {0} who is majoring in {1}.'.format(self.myStudentType,self.myMajor))
        print(self.myWhy, end='')
        print('\n{0}.'.format(self.myMolBio,))
        print(self.PriorCompExp)


'''defining variables that are then put into a new Person class'''
name = 'Toby Grief'
username = 'tgrief'
major = 'BME'
studentType = 'Junior'
why = 'I\'m taking this class because its required by my major and I am hoping to use the python skills I learn in this class later in my career.'
MolBio = 'I\'m interested in learning more about proteomics and CRISPR gene editing'
PriorCompExp = 'I took python and java in high school and I have not coded in a while.'
toby = Person(name,username,major,studentType,why,MolBio,PriorCompExp)

'''executes the program'''
toby.introduce()
```

```python
#!/usr/bin/env python3
# Name: Toby Grief
# Group Members: Nathan Perez, Theodore Shen, Bharath Karintha, Nate Cruz
'''Counting DNA Letters
countNucleotides'''

class dnaString (str):
''' This class counts the number of Nucleotides sent by coolString'''
    def length (self): #Counts the length of the dna sequence
        return (len(self))

    def getA (self):#counts the number As
        num_A = self.count("A")
        return num_A

    def getT (self):#counts the number Ts
        num_T = self.count("T")
        return num_T

    def getG (self):#coutns the number Gs
        num_G = self.count("G")
        return num_G

    def getC (self):#counts the number of Cs
        num_C = self.count("C")
        return num_C

dna = input("Enter a dna sequence: ")#User inputs a sequence
upperDNA = dna.upper()#All letters in the string are then uppercase
coolString = dnaString(upperDNA)#this creates the object and stores user info

print ("Your sequence is {} nucleotides long with the following "
       "breakdown of bases:"
       ""
       " number As = {} number Cs = {} number Gs = {} number Ts = {}".format(
    coolString.length(),#Takes the object coolString and counts the number of nucleotides
    coolString.getA(),#Shows the number of As
    coolString.getC(),#Shows the number of Cs
    coolString.getG(),#Shows the number of Gs
    coolString.getT()) )#Shows the number of Ts
```
